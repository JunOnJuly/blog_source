```python
from matplotlib import pyplot as plt
```

```python
fig, axes = plt.subplots(1, 2, figsize = (12, 4))
                         
axes[0].plot([5, 5, 1], [0, 10, 6], 'o-')
axes[0].plot([3, 3], [0, 8], 'o-')
axes[0].plot([1, 1], [0, 5], 'o-')
                         
axes[1].plot([5, 5, 1], [0, 10, 6], 'o-')
axes[1].plot([3, 3], [0, 8], 'o-')
axes[1].plot([1, 1], [0, 9], 'o-')
                         
plt.savefig('under.png', dpi=600)
```
