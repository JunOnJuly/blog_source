```python
from matplotlib import pyplot as plt
import os
from PIL import Image
```

```python
# 점을 기록할 클래스
class Point:
    def __init__(self, x, y):
    	# 좌표
        self.x = x
        self.y = y


# 선분을 기록할 클래스
class Line:
    def __init__(self, pt_1, pt_2):
        # 양 끝점
        self.pt_1 = pt_1
        self.pt_2 = pt_2


# ccw
def ccw(sight, building_top):
    # 외적을 위한 벡터
    ccw_vec = [ sight.pt_1, sight.pt_2, building_top, sight.pt_1 ]
    # 외적 값
    ccw_value = 0
    # 계산
    for i in range(3):
        ccw_value += ccw_vec[i].x * ccw_vec[i + 1].y
        ccw_value -= ccw_vec[i + 1].x * ccw_vec[i].y
        
    # 음수면 시계방향(-1), 양수면 반시계방향(1), 0 이면 같은 직선상(0) 
    return 1 if ccw_value > 0 else 0 if ccw_value == 0 else -1


# 현재 건물에서 보이는 건물의 수를 세는 함수
def count_build(now_build, next_build, buildings, dir, build_num, N):
    global img_cnt
    # 사선을 이을 다음 건물이 인덱스를 벗어나면
    if next_build < 0 or next_build >= N: return build_num
    # 현재 위치
    now_pt = Point(now_build, buildings[now_build])
    # 시선을 이을 다음 위치
    next_pt = Point(next_build, buildings[next_build])
    # 시선 선분
    sight = Line(now_pt , next_pt)
    # 타겟 인덱스
    target_build = next_build

    # 시각화 ------------------------------
    axes[0].cla()
    axes[0].set_xlim([-0.5, 15.5])
    axes[0].set_ylim([-0.5, 10.5])
    axes[0].set_xticks(range(0, 15))
    axes[0].set_yticks(range(0, 11))

    for idx in range(N):
        axes[0].plot([idx, idx], [0, buildings[idx]], color=bar_colors[idx], linewidth=5)
    axes[0].plot([now_pt.x, next_pt.x], [now_pt.y, next_pt.y], color=bar_colors[now_pt.x], linewidth=2)

    axes[1].cla()
    axes[1].set_xlim([-0.5, 10.5])
    axes[1].set_ylim([-0.5, 10.5])
    axes[1].set_xticks([])
    axes[1].set_yticks([])
    axes[1].axis("off")

    axes[1].text(0, 0, f'max_build : {max_build}', fontsize=16)
    axes[1].text(0, 0.5, f'build_num : {build_num}', fontsize=16)
    axes[1].text(0, 1, f'now_pt : {[now_pt.x, now_pt.y]}', fontsize=16)
    axes[1].text(0, 1.5, f'next_pt : {[next_pt.x, next_pt.y]}', fontsize=16)

    plt.savefig(f'../IMAGES/1027/{img_cnt}.png', dpi=600)
    img_cnt += 1
    # --------------------------------------
                                              
    # 순회
    while True:
        # 왼쪽 방향이면
        if not dir:
            # 방향을 비교할 인덱스, 포인트
            target_build -= 1
            # 방향을 비교할 인덱스가 범위를 넘으면 리턴
            if target_build < 0: return build_num
        # 오른쪽 방향이면
        else:
            # 방향을 비교할 인덱스, 포인트
            target_build += 1
            # 방향을 비교할 인덱스가 범위를 넘으면 리턴
            if target_build >= N: return build_num
        
        # 방향을 비교할 포인트
        target_pt = Point(target_build, buildings[target_build])

        # 시각화 ------------------------------
        axes[0].cla()
        axes[0].set_xlim([-0.5, 15.5])
        axes[0].set_ylim([-0.5, 10.5])
        axes[0].set_xticks(range(0, 15))
        axes[0].set_yticks(range(0, 11))

        for idx in range(N):
            axes[0].plot([idx, idx], [0, buildings[idx]], color=bar_colors[idx], linewidth=5)
        axes[0].plot([now_pt.x, next_pt.x], [now_pt.y, next_pt.y], color=bar_colors[now_pt.x], linewidth=2)
        axes[0].plot([next_pt.x, target_pt.x], [next_pt.y, target_pt.y], color=bar_colors[now_pt.x], linewidth=2)

        axes[1].cla()
        axes[1].set_xlim([-0.5, 10.5])
        axes[1].set_ylim([-0.5, 10.5])
        axes[1].set_xticks([])
        axes[1].set_yticks([])
        axes[1].axis("off")
        
        axes[1].text(0, 0, f'max_build : {max_build}', fontsize=16)
        axes[1].text(0, 0.5, f'build_num : {build_num}', fontsize=16)
        axes[1].text(0, 1, f'now_pt : {[now_pt.x, now_pt.y]}', fontsize=16)
        axes[1].text(0, 1.5, f'next_pt : {[next_pt.x, next_pt.y]}', fontsize=16)
        axes[1].text(0, 2, f'target_pt : {[target_pt.x, target_pt.y]}', fontsize=16)
        
        plt.savefig(f'../IMAGES/1027/{img_cnt}.png', dpi=600)
        img_cnt += 1
        # --------------------------------------
        
        # ccw 방향
        ccw_dir = ccw(sight, target_pt)
        # 왼쪽 방향이면
        if not dir:
            # 시계 방향이면
            if ccw_dir == -1:
                # 시각화 ------------------------------
                axes[0].cla()
                axes[0].set_xlim([-0.5, 15.5])
                axes[0].set_ylim([-0.5, 10.5])
                axes[0].set_xticks(range(0, 15))
                axes[0].set_yticks(range(0, 11))
            
                for idx in range(N):
                    axes[0].plot([idx, idx], [0, buildings[idx]], color=bar_colors[idx], linewidth=5)
                axes[0].plot([now_pt.x, next_pt.x], [now_pt.y, next_pt.y], color='blue', linewidth=2)
                axes[0].plot([next_pt.x, target_pt.x], [next_pt.y, target_pt.y], color='blue', linewidth=2)

                axes[1].cla()
                axes[1].set_xlim([-0.5, 10.5])
                axes[1].set_ylim([-0.5, 10.5])
                axes[1].set_xticks([])
                axes[1].set_yticks([])
                axes[1].axis("off")
                
                axes[1].text(0, 0, f'max_build : {max_build}', fontsize=16)
                axes[1].text(0, 0.5, f'build_num : {build_num}', fontsize=16)
                axes[1].text(0, 1, f'now_pt : {[now_pt.x, now_pt.y]}', fontsize=16)
                axes[1].text(0, 1.5, f'next_pt : {[next_pt.x, next_pt.y]}', fontsize=16)
                axes[1].text(0, 2, f'target_pt : {[target_pt.x, target_pt.y]}', fontsize=16)
                
                plt.savefig(f'../IMAGES/1027/{img_cnt}.png', dpi=600)
                img_cnt += 1
                # --------------------------------------
        
                # 보이는 건물 + 1
                build_num += 1
                build_num = count_build(now_build, target_build, buildings, dir, build_num, N)
                break
        # 오른쪽 방향이면
        else:
            # 반시계 방향이면
            if ccw_dir == 1:
                # 시각화 ------------------------------
                axes[0].cla()
                axes[0].set_xlim([-0.5, 15.5])
                axes[0].set_ylim([-0.5, 10.5])
                axes[0].set_xticks(range(0, 15))
                axes[0].set_yticks(range(0, 11))
            
                for idx in range(N):
                    axes[0].plot([idx, idx], [0, buildings[idx]], color=bar_colors[idx], linewidth=5)
                axes[0].plot([now_pt.x, next_pt.x], [now_pt.y, next_pt.y], color='blue', linewidth=2)
                axes[0].plot([next_pt.x, target_pt.x], [next_pt.y, target_pt.y], color='blue', linewidth=2)

                axes[1].cla()
                axes[1].set_xlim([-0.5, 10.5])
                axes[1].set_ylim([-0.5, 10.5])
                axes[1].set_xticks([])
                axes[1].set_yticks([])
                axes[1].axis("off")
                
                axes[1].text(0, 0, f'max_build : {max_build}', fontsize=16)
                axes[1].text(0, 0.5, f'build_num : {build_num}', fontsize=16)
                axes[1].text(0, 1, f'now_pt : {[now_pt.x, now_pt.y]}', fontsize=16)
                axes[1].text(0, 1.5, f'next_pt : {[next_pt.x, next_pt.y]}', fontsize=16)
                axes[1].text(0, 2, f'target_pt : {[target_pt.x, target_pt.y]}', fontsize=16)
                
                plt.savefig(f'../IMAGES/1027/{img_cnt}.png', dpi=600)
                img_cnt += 1
                # --------------------------------------
                
                # 보이는 건물 + 1
                build_num += 1
                build_num = count_build(now_build, target_build, buildings, dir, build_num, N)
                break
    
    return build_num


# 시각화 옵션 ---------------------------
fig, axes = plt.subplots(1, 2, figsize=(8,6), gridspec_kw={'width_ratios': [3, 1]})
axes[0].set_xlim([-0.5, 15.5])
axes[0].set_ylim([-0.5, 10.5])
axes[0].set_xticks(range(0, 15))
axes[0].set_yticks(range(0, 11))

axes[1].set_xlim([-0.5, 10.5])
axes[1].set_ylim([-0.5, 10.5])
axes[1].set_xticks([])
axes[1].set_yticks([])
axes[1].axis("off")

c_base = 'lightgreen'
c_effect = 'olivedrab'
c_false = 'springgreen'
c_true = 'darkseagreen'
c_block = 'darkgreen'

bar_colors = ['lightcoral', 'bisque', 'tan', 'orange', 'khaki', 'olive', 'yellowgreen',
             'palegreen', 'mediumaquamarine', 'aquamarine', 'paleturquoise', 'teal', 'deepskyblue',
             'steelblue', 'slategray', 'lightsteelblue', 'navy', 'mediumslateblue', 'blueviolet', 'plum', 'fuchsia',
             'orchid', 'hotpink', 'pink']
# --------------------------------------

# 입력
# N = int(input())
N = 15
# 건물 높이를 기록할 벡터
# buildings = list(map(int, input().split()))
buildings = [1, 5, 3, 2, 6, 3, 2, 6, 4, 2, 5, 7, 3, 1, 5]

# 시각화 ------------------------------
# 이미지 카운트
global img_cnt
img_cnt = 0
for idx in range(N):
    axes[0].plot([idx, idx], [0, buildings[idx]], color=bar_colors[idx], linewidth = 5)
plt.savefig(f'../IMAGES/1027/{img_cnt}.png', dpi=600)
img_cnt += 1
# --------------------------------------

# 보이는 건물의 최댓값
global max_build
max_build = 0
# 건물 위치 순회
for i in range(N):
    # 현재 건물
    now_build = i
    # 현재 건물에서 보이는 건물의 수 ( 바로 옆 건물들은 그냥 보임 )
    watchable_build = 0 if i==0 and i==N-1 else 1 if i==0 or i==N else 2
    # 현재 건물 왼쪽 건물
    left_build = now_build - 1
    # 왼쪽 방향
    left_dir = False
    # 현재 건물에서 왼쪽 탐색
    watchable_build = count_build(now_build, left_build, buildings, left_dir, watchable_build, N)
    # 현재 건물 오른쪽 건물
    right_build = now_build + 1
    # 오른쪽 방향
    right_dir = True
    # 현재 건물에서 오른쪽 탐색
    watchable_build = count_build(now_build, right_build, buildings, right_dir, watchable_build, N)
    # 최댓값 최신화
    max_build = max(max_build, watchable_build)

# 시각화 ------------------------------
img_list = ['../IMAGES/1027' + '/' + img for img in os.listdir('../IMAGES/1027')]
img_list.sort(key=lambda x: int(x.split('/')[3][:-4]))
images = [Image.open(img) for img in img_list]

gif = images[0]
gif.save('../IMAGES/GIFS/1027.gif', save_all=True, append_images=images[1:], loop=1, duration=500)
# -------------------------------------
```
