```python
from matplotlib import pyplot as plt
from functools import cmp_to_key
import random
from IPython import display
from time import sleep
```

```python
def ccw(point_1, point_2, point_3):
    points = [point_1, point_2, point_3, point_1]
    ccw_sum = 0
    for i in range(3):
        ccw_sum += points[i][0] * points[i+1][1]
        ccw_sum -= points[i+1][0] * points[i][1]
    return ccw_sum
    
def sort_clock_wise(point_1, point_2):
    origin_point = [point_1[2], point_1[3], None, None]
    ccw_points = ccw(origin_point, point_1, point_2)
    dist_o1 = pow((point_1[3] - point_1[1]), 2) + pow((point_1[2] - point_1[0]), 2)
    dist_o2 = pow((point_2[3] - point_2[1]), 2) + pow((point_2[2] - point_2[0]), 2)
    if ccw_points < 0:
        return -1
    elif ccw_points == 0 and dist_o1 < dist_o2:
        return -1
    else:
        return 1
    
def graham_scan(points):
    stk = points[:3]
    idx = 3
    
    max_x_points = max([pt[0] for pt in points])
    min_y_points = min([pt[1] for pt in points])
    
    while (True):
        plt.clf()
        plt.text(max_x_points+15, min_y_points, "stack", fontsize=20)
        for i in range(len(stk)):
            plt.text(max_x_points+15, min_y_points+(15*(i+1)), f"[{stk[i][0]}, {stk[i][1]}]", fontsize=20)
        point_1 = stk[-2][:2]
        point_2 = stk[-1][:2]
        point_1.extend(stk[-3][:2])
        point_2.extend(stk[-3][:2])
        
        plt.xlim(min(xs)-5, max(xs)+40)
        plt.gca().set_axis_off()
        plt.plot([stk[-3][0], stk[-2][0]], [stk[-3][1], stk[-2][1]], 'bo-', linewidth=5)
        display.clear_output(wait=True)
        display.display(plt.gcf()) 
        sleep(0.2)
        
        plt.xlim(min(xs)-5, max(xs)+40)
        plt.gca().set_axis_off()
        plt.plot([stk[-2][0], stk[-1][0]], [stk[-2][1], stk[-1][1]], 'bo-', linewidth=5)
        display.clear_output(wait=True)
        display.display(plt.gcf()) 
        sleep(0.2)
        
        if sort_clock_wise(point_1, point_2) == 1:
            plt.xlim(min(xs)-5, max(xs)+40)
            plt.gca().set_axis_off()
            plt.plot([stk[-3][0], stk[-2][0], stk[-1][0]], [stk[-3][1], stk[-2][1], stk[-1][1]], 'r', linewidth=5)
            display.clear_output(wait=True)
            display.display(plt.gcf()) 
            sleep(0.5)
            del(stk[-2])
        else:
            point_3 = stk[-3]
            if ccw(point_3, point_1, point_2) == 0:
                del(stk[-2])
            else:
                if idx >= len(points):
                    break
                stk.append(points[idx])
                idx += 1
        if len(stk)<= 2:
            if idx >= len(points):
                break
            stk.append(points[idx])
            idx += 1
    return stk
```

```python
N = random.randint(50, 60)

plt.figure(figsize=(20, 6))
xs = random.sample(range(int(-1.5*N), int(1.5*N + 1)), N)
ys = random.sample(range(int(-1.5*N), int(1.5*N + 1)), N)
plt.xlim(min(xs)-5, max(xs)+40)
plt.gca().set_axis_off()
plt.plot(xs, ys, 'k.')

display.clear_output(wait=True)
display.display(plt.gcf()) 
sleep(1)

points = [[x, y] for x, y in zip(xs, ys)]
points.sort(key=lambda x:(x[0], x[1]))
plt.xlim(min(xs)-5, max(xs)+40)
plt.gca().set_axis_off()
plt.plot([x[0] for x in points], [y[1] for y in points], 'ko-')

display.clear_output(wait=True)
display.display(plt.gcf()) 
sleep(1)
plt.clf()

for i in range(N):
    points[i].extend(points[0][:2])

points.sort(key=cmp_to_key(sort_clock_wise))
plt.xlim(min(xs)-5, max(xs)+40)
plt.gca().set_axis_off()
plt.plot([x[0] for x in points], [y[1] for y in points], 'bo-')

display.clear_output(wait=True)
display.display(plt.gcf()) 
sleep(1)

convex_hull = graham_scan(points)
plt.xlim(min(xs)-5, max(xs)+40)
plt.gca().set_axis_off()
plt.plot([convex_hull[0][0], convex_hull[-1][0]], [convex_hull[0][1], convex_hull[-1][1]], 'bo-', linewidth=5)

display.clear_output(wait=True)
display.display(plt.gcf()) 
```
