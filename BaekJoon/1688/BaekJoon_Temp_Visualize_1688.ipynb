```python
from matplotlib import pyplot as plt
import random
from time import sleep
from IPython import display
```

```python
class Point:
    def __init__(self, x, y):
        self.x = x
        self.y = y

def ccw(point_1, point_2, point_3):
    points = [ point_1, point_2, point_3, point_1 ]
    ccw_sum = 0;
    for i in range(3):
        ccw_sum += points[i].x * points[i + 1].y
        ccw_sum -= points[i].y * points[i + 1].x

    return ccw_sum

class Line:
    def __init__(self, point_1, point_2):
        self.point_1 = point_1
        self.point_2 = point_2

def is_cross(line_1, line_2):
    ccw_1 = ccw(line_1.point_1, line_1.point_2, line_2.point_1)
    ccw_2 = ccw(line_1.point_1, line_1.point_2, line_2.point_2)
    ccw_3 = ccw(line_2.point_1, line_2.point_2, line_1.point_1)
    ccw_4 = ccw(line_2.point_1, line_2.point_2, line_1.point_2)
    
    result_1 = ccw_1 * ccw_2
    result_2 = ccw_3 * ccw_4

    if (result_1 < 0 and result_2 < 0): return True
    elif (result_1 > 0 and result_2 > 0): return False
    elif ((result_1 == 0 and result_2 < 0) or (result_1 < 0 and result_2 == 0)): return True
    elif ((result_1 == 0 and result_2 > 0) or (result_1 > 0 and result_2 == 0)): return False
    elif ((result_1 > 0 and result_2 < 0) or (result_1 < 0 and result_2 > 0)): return False
    else:
        if ((((line_1.point_1.x > line_2.point_1.x and line_1.point_2.x > line_2.point_1.x) and
            (line_1.point_1.x > line_2.point_2.x and line_1.point_2.x > line_2.point_2.x)) or
            ((line_1.point_1.y > line_2.point_1.y and line_1.point_2.y > line_2.point_1.y) and
                (line_1.point_1.y > line_2.point_2.y and line_1.point_2.y > line_2.point_2.y))) or

            (((line_2.point_1.x > line_1.point_1.x and line_2.point_2.x > line_1.point_1.x) and
                (line_2.point_1.x > line_1.point_2.x and line_2.point_2.x > line_1.point_2.x)) or
                ((line_2.point_1.y > line_1.point_1.y and line_2.point_2.y > line_1.point_1.y) and
                    (line_2.point_1.y > line_1.point_2.y and line_2.point_2.y > line_1.point_2.y)))): return False
        else: return True

line_width = 3
N = 13
points_lst = [[14, 8], [10, 5], [10, 10], [5, 10], [15, 15], [3, 13], [2, 16], [1, 12], [3, 8], [6, 8], [8, 3], [12, 1], [15, 2], [14, 8]]
people_lst = [[3, 10], [4, 2], [8, 3], [8, 10], [2, 8], [12, 4]]
points = [Point(x, y) for x, y in points_lst]
people = [Point(x, y) for x, y in people_lst]

plt.plot([x for x, y in points_lst], [y for x, y in points_lst], 'ko-', linewidth=line_width)
plt.plot([x for x, y in people_lst], [y for x, y in people_lst], 'mo')

plt.xlim(0, 16)
plt.ylim(0, 17)
display.clear_output(wait=True)
display.display(plt.gcf()) 
sleep(1)

for i in range(len(people)):
    person = people[i];
    max_x = 21;
    max_y = person.y + 1;
    person_line = Line(person, Point(max_x, max_y))
    
    plt.clf()
    plt.plot([person.x, max_x], [person.y, max_y], 'k--')
    plt.plot([x for x, y in points_lst], [y for x, y in points_lst], 'ko-', linewidth=line_width)
    plt.plot([x for x, y in people_lst], [y for x, y in people_lst], 'mo')
    plt.plot(person.x, person.y, 'go')
    plt.xlim(0, 16)
    plt.ylim(0, 17)
    display.clear_output(wait=True)
    display.display(plt.gcf()) 
    sleep(1)
    
    for j in range(N):
        if (person.x >= min(points[j].x, points[j + 1].x) and person.x <= max(points[j].x, points[j + 1].x) and
            person.y >= min(points[j].y, points[j + 1].y) and person.y <= max(points[j].y, points[j + 1].y) and
            ccw(points[j], points[j + 1], person) == 0):
            
            plt.clf()
            plt.plot([person.x, max_x], [person.y, max_y], 'k--')
            plt.plot([x for x, y in points_lst], [y for x, y in points_lst], 'ko-', linewidth=line_width)
            plt.plot([x for x, y in people_lst], [y for x, y in people_lst], 'mo')
            plt.plot(person.x, person.y, 'ro')
            plt.plot([points[j].x, points[j+1].x], [points[j].y, points[j+1].y], 'r', linewidth=line_width)
            plt.xlim(0, 16)
            plt.ylim(0, 17)
            display.clear_output(wait=True)
            display.display(plt.gcf()) 
            sleep(1)
            
            break;

        else:
            barrier_line = Line(points[j], points[j + 1])
            
            plt.clf()
            plt.plot([person.x, max_x], [person.y, max_y], 'k--')
            plt.plot([x for x, y in points_lst], [y for x, y in points_lst], 'ko-', linewidth=line_width)
            plt.plot([x for x, y in people_lst], [y for x, y in people_lst], 'mo')
            plt.plot(person.x, person.y, 'go')
            plt.plot([points[j].x, points[j+1].x], [points[j].y, points[j+1].y], 'g', linewidth=line_width)
            plt.xlim(0, 16)
            plt.ylim(0, 17)
            display.clear_output(wait=True)
            display.display(plt.gcf()) 
            sleep(0.2)
            
            if (is_cross(person_line, barrier_line)):           
                plt.clf()
                plt.plot([x for x, y in points_lst], [y for x, y in points_lst], 'ko-', linewidth=line_width)
                plt.plot([x for x, y in people_lst], [y for x, y in people_lst], 'mo')
                plt.plot([person.x, max_x], [person.y, max_y], 'k--')
                plt.plot(person.x, person.y, 'go')
                plt.plot([points[j].x, points[j+1].x], [points[j].y, points[j+1].y], 'b', linewidth=line_width)
                plt.xlim(0, 16)
                plt.ylim(0, 17)
                display.clear_output(wait=True)
                display.display(plt.gcf()) 
                sleep(0.2)
            else:
                plt.clf()
                plt.plot([x for x, y in points_lst], [y for x, y in points_lst], 'ko-', linewidth=line_width)
                plt.plot([x for x, y in people_lst], [y for x, y in people_lst], 'mo')
                plt.plot([person.x, max_x], [person.y, max_y], 'k--')
                plt.plot(person.x, person.y, 'go')
                plt.plot([points[j].x, points[j+1].x], [points[j].y, points[j+1].y], 'r', linewidth=line_width)
                plt.xlim(0, 16)
                plt.ylim(0, 17)
                display.clear_output(wait=True)
                display.display(plt.gcf()) 
                sleep(0.2)
```
